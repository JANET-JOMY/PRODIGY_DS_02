```python
#finding the car which have covered the maximum distance
max_distance_index = df['km_driven'].idxmax()
car_with_max_distance = df.loc[max_distance_index]
print("Car with Maximum Distance Covered:")
print(car_with_max_distance)
```

```python
#car with maximum selling price
max_sellingprice = df['selling_price'].idxmax()
car_with_max_sellingprice = df.loc[max_sellingprice]
print("Car with Maximum selling price:")
print(car_with_max_sellingprice)
```

```python
df["name_2"] = df.name.apply(lambda x : ' '.join(x.split(' ')[:1]))
df['name_2']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(data=df,x="name_2",palette="CMRmap")
plt.xticks(rotation=90)
plt.xlabel("Name",fontsize=10,color="blue")
plt.ylabel("Name",fontsize=10,color="blue")
plt.title("NAME COUNT",color="blue")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt  
import pandas as pd 
sns.countplot(x='fuel',hue='transmission',data=df)
plt.title('Types of fuel and transmission data')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='seller_type', data=df)
plt.title('Count of Cars by Seller Type')
plt.xlabel('Seller Type')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='transmission', data=df)
plt.title('Count of Cars by Transmission Type')
plt.xlabel('Transmission Type')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='owner', data=df)
plt.title('Count of Cars by Owner Type')
plt.xlabel('Owner Type')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.subplot(2, 2, 1)
sns.histplot(df['selling_price'], kde=True)
plt.title('Distribution of Selling Price')

plt.subplot(2, 2, 2)
sns.histplot(df['km_driven'], kde=True)
plt.title('Distribution of Kilometers Driven')
```

```python
labels = df["year"][:40].value_counts().index
sizes = df["year"][:40].value_counts()
colors = ['#ff9999','#66b3ff','#99ff89','#ffcc99',"red","yellow"]
plt.figure(figsize = (6,8))
plt.pie(sizes, labels=labels, rotatelabels=False, autopct='%1.1f%%',colors=colors,shadow=True, startangle=45)
plt.title('Year',color = 'red',fontsize = 15)
plt.show()
```

```python
sns.countplot(data=df,x="year",palette="icefire")
plt.xticks(rotation=90)
plt.xlabel("YEAR",fontsize=10,color="black")
plt.ylabel("COUNT",fontsize=10,color="RED")
plt.title("YEAR COUNT",color="blue")
plt.show()
```

```python
plt.figure(figsize=(15, 8))

plt.subplot(2, 2, 3)
sns.countplot(x='fuel', data=df)
plt.title('Count of Cars by Fuel Type')

plt.subplot(2, 2, 4)
sns.countplot(x='owner', data=df)
plt.title('Count of Cars by Owner Type')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='selling_price', y='km_driven', data=df)
plt.title('Selling Price vs Kilometers Driven')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='fuel', y='selling_price', data=df)
plt.title('Distribution of Selling Price by Fuel Type')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='fuel', y='selling_price', data=df)
plt.title('Distribution of Selling Price by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Selling Price')
plt.show()
```

```python
sns.violinplot(data=df, x="year", y="fuel",hue="transmission")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='year', y='selling_price', data=df)
plt.title('Trend of Selling Price Over Year')
plt.xlabel('Year')
plt.ylabel('Selling Price')
plt.show()
```

```python
# Correlation matrix
correlation_matrix = df[['selling_price', 'km_driven', 'year']].corr()

# Heatmap of correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
# Ridgeline plot of selling_price by fuel type
plt.figure(figsize=(15, 6))
sns.kdeplot(data=df, x='selling_price', hue='fuel', fill=True)
plt.title('Ridgeline Plot of Selling Price by Fuel Type')
plt.xlabel('Selling Price')
plt.ylabel('Density')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Correlation matrix
correlation_matrix = df[['selling_price', 'km_driven', 'year']].corr()

# Correlogram
sns.pairplot(df[['selling_price', 'km_driven', 'year']])
plt.suptitle('Correlogram', y=1.02)
plt.figure(figsize=(15, 6))
plt.show()
```

It was found:
Cars with fuel type "petrol" has high rate of transmission : "manual".
The pie chart indicates that most of the cars are manufactured in the year 2017.
The best selling car is the Maruti. Next comes Hyundai.
Cars are usually manual.
We can say that vehicles are generally sold first-hand, followed by second-hand vehicles.
Most of the sales are made individually.
There are very few outliers in the data.
Line plot indicates that selling price is increasing as years increase.
